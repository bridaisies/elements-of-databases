```python
# cross dataset queries
```

```python
# find movies from both datasets that are in the genre war
```

```python
%%bigquery --use_rest_api
select distinct t.Title, t.Genre, n.genres
from kaggle_refined.Title_genre_Dataflow as t
join (select primaryTitle, genres
from imdb_refined.Title_Basics as b
join imdb_refined.Genre_Dataflow as g
on b.tconst = g.tconst) as n 
on n.genres = t.Genre
where t.Genre = 'war' and n.genres = 'war'
```

```python
# find the average ratings of Mani Ratnam movies
```

```python
%%bigquery 
select distinct primaryTitle, averageRating from imdb_refined.Title_Basics
where primaryTitle in
(SELECT Title
FROM kaggle_refined.Title_director_Dataflow
WHERE Director = 'Mani Ratnam')
```

```python
# find all the primary professions of Bollywood "actors" (Data Studio visualization)
```

```python
%%bigquery 
create view insights.v_bollywoodPrimaryProf as
select distinct Name, p.primaryProfession from kaggle_refined.Bolly_actors a
join imdb_refined.Name_Basics b on a.Name = b.primaryName
join imdb_refined.Primary_Professions_Dataflow p on p.nconst = b.nconst
where primaryProfession != ''
group by Name, p.primaryProfession
```

```python
# collective info about living Bollywood actresses
```

```python
%%bigquery
select primaryName, birthYear, Height_in_cm_ from imdb_refined.Name_Basics n
join kaggle_refined.Bolly_actress b on n.primaryName = b.Name
where deathYear is NULL and birthYear is NOT NULL
order by Height_in_cm desc
limit 25
```

```python
# count most common genres in Hollywood vs Bollywood (Data Studio visualization)
```

```python
%%bigquery
create view insights.v_hollywoodGenres as
select genres, count(genres) as genre_count from imdb_refined.Genre_Dataflow
where genres is not null
group by genres
order by genre_count desc
limit 10
```

```python
%%bigquery 
create view insights.v_bollywoodGenres as
select Genre, count(Genre) as genre_count from kaggle_refined.Title_genre_Dataflow
where Genre is not null and Genre != 'unknown'
group by Genre
order by genre_count desc
limit 10
```

```python
# see if actor height is correlated to popularity (Data Studio visualization)
```

```python
%%bigquery 
create view insights.v_heightAndPopularity as
select primaryName, count(knownForTitles) as number_movies, b.Height_in_cm_ from imdb_refined.Known_For_Dataflow k
join imdb_refined.Name_Basics n on n.nconst = k.nconst
join kaggle_refined.Bolly_actors b on n.primaryName = b.Name
where n.nconst in (select distinct nconst from kaggle_refined.Bolly_actors b
join imdb_refined.Name_Basics n on n.primaryName = b.Name)
group by primaryName, b.Height_in_cm_
order by b.Height_in_cm_ desc
```
