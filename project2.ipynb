```python
# See the category that actor #194 acts in through joining these tables
```

```python
!psql -c "select actor_id, category_id from pagila.film_actor inner join pagila.film_category on pagila.film_actor.film_id = pagila.film_category.film_id where actor_id = 194 order by actor_id limit 5"
```

```python
# Join the tables in a way that shows the film IDs of the movies that Jennifer acted in
```

```python
!psql -c "select first_name, last_name, film_id from pagila.actor inner join pagila.film_actor on pagila.actor.actor_id = pagila.film_actor.actor_id where first_name = 'JENNIFER' order by film_id limit 5"
```

```python
# Join the tables to show the title and length from one table, and the corresponding language name from another table
```

```python
!psql -c "select title, name, length from pagila.film inner join pagila.language on pagila.film.language_id = pagila.language.language_id where length < 48 order by length"
```

```python
# Use a join between 2 tables to find which city is in which district, and order this by country ID
```

```python
!psql -c "select city, district, country_id from pagila.address left outer join pagila.city on pagila.address.city_id = pagila.city.city_id where country_id < 5 order by country_id"
```

```python
# Get a table of customers and the city/ districts they are from, using a right outer join
```

```python
!psql -c "select customer_id, first_name, last_name, city_id, district from pagila.customer right outer join pagila.address on pagila.customer.address_id = pagila.address.address_id where customer_id < 10 order by city_id"
```

```python
# Do a full join to get staff information and return their manager ID number, respectively
```

```python
!psql -c "select staff_id, first_name, last_name, manager_staff_id from pagila.staff full outer join pagila.store on pagila.staff.store_id = pagila.store.store_id where staff_id < 3 order by manager_staff_id"
```
