```python
# make new table to get rid of 1 to many relationship between tconst and genre (comma-separated values will be separated for next milestone)
```

```python
%%bigquery
CREATE TABLE imdb_refined.Genre AS
SELECT tconst, genres
FROM imdb_staging.Title_Basics;
```

```python
# make new table to get rid of 1 to many relationship between nconst and knownForTitles (comma-separated values will be separated for next milestone)
```

```python
%%bigquery
CREATE TABLE imdb_refined.Known_For AS
SELECT nconst, knownForTitles
FROM imdb_staging.Name_Basics;
```

```python
# make new table to get rid of 1 to many relationship between nconst and primaryProfession (comma-separated values will be separated for next milestone)
```

```python
%%bigquery
CREATE TABLE imdb_refined.Primary_Professions AS
SELECT nconst, primaryProfession
FROM imdb_staging.Name_Basics;
```

```python
# make new table to get rid of 1 to many relationship between nconst and characters (comma-separated values will be separated for next milestone)
```

```python
%%bigquery
CREATE TABLE imdb_refined.Characters AS
SELECT tconst, nconst, characters
FROM imdb_staging.Title_Principals;
```

```python
# combine the title_basics and title_akas tables
```

```python
%%bigquery
CREATE TABLE imdb_refined.Title_Basics AS 
SELECT b.tconst, b.titleType, b.primaryTitle, b.isAdult, b.startYear, b.endYear, CAST(b.runtimeMinutes as INT64) as runtimeMin, r.averageRating, r.numVotes, a.region, a.language, a.title
FROM imdb_staging.Title_Basics b
FULL OUTER JOIN imdb_staging.Title_Akas a ON b.tconst = a.titleId
FULL OUTER JOIN imdb_staging.Title_Ratings r on r.tconst = b.tconst;
```

```python
# update name_basics table to get rid of profession and knownFor columns
```

```python
%%bigquery
CREATE TABLE imdb_refined.Name_Basics AS
SELECT nconst, primaryName, birthYear, deathYear
FROM imdb_staging.Name_Basics;
```

```python
# resolve the 1:m relationship in the crews table between tconst and directors, and tconst and writers (will separate comma-separated values for the next milestone)
```

```python
%%bigquery
CREATE TABLE imdb_refined.Directors AS
SELECT tconst, directors
FROM imdb_staging.Title_Crew;
```

```python
%%bigquery
CREATE TABLE imdb_refined.Writers AS
SELECT tconst, writers
FROM imdb_staging.Title_Crew;
```

```python
# import episode table with no alterations
```

```python
%%bigquery
CREATE TABLE imdb_refined.Title_Episode AS
SELECT * FROM imdb_staging.Title_Episode;
```

```python
# import the principals table without the ordering and characters columns
```

```python
%%bigquery
CREATE TABLE imdb_refined.Title_Principals AS
SELECT tconst, nconst, category, job
FROM imdb_staging.Title_Principals;
```

```python
# find the movies that had a higher than average rating
```

```python
%%bigquery
select tconst, primaryTitle, averageRating
from imdb_refined.Title_Basics
where Title_Basics.averageRating > (select AVG(averageRating) from imdb_refined.Title_Basics)
order by averageRating desc;
```

```python
# find all of the tallest Bollywood actors
```

```python
%%bigquery
select Name, Height_in_cm_
from kaggle_refined.Bolly_actors
where Height_in_cm_ = (select max(Height_in_cm_) from kaggle_refined.Bolly_actors)
order by Name
```

```python
# find all the Bollywood actresses who are the same height as male directors who are also actors
```

```python
%%bigquery
select Name, Height_in_cm_ 
from kaggle_refined.Bolly_actress
where Height_in_cm_ IN
(select Height_in_cm_
FROM kaggle_refined.Title_director as m
JOIN kaggle_refined.Bolly_actors as a
ON m.Director = a.Name)
LIMIT 10
```

```python
# find the actors who are not also directors
```

```python
%%bigquery
select primaryName 
from imdb_refined.Name_Basics
where primaryName not in 
(select primaryName
from imdb_refined.Name_Basics n
inner join imdb_refined.Primary_Professions p on n.nconst = p.nconst
WHERE primaryProfession = 'actor')
LIMIT 10
```

```python
# find the regions that received over 1 million total votes
```

```python
%%bigquery
select region, sum(numVotes) as total_votes
from imdb_refined.Title_Basics
group by region
having total_votes > 1000000
```

```python
#CREATE TWO VIEWS
```

```python
%%bigquery
create view insights.v_higherThanAvgRating as
select tconst, primaryTitle, averageRating
from imdb_refined.Title_Basics
where Title_Basics.averageRating > (select AVG(averageRating) from imdb_refined.Title_Basics)
order by averageRating desc;
```

```python
%%bigquery
create view insights.v_RegionVotes as
select region, sum(numVotes) as total_votes
from imdb_refined.Title_Basics
group by region
having total_votes > 1000000;
```
