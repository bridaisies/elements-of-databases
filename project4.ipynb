```python
import pandas as pd
from google.cloud import firestore
db = firestore.Client()
```

```python
batch = db.batch()

df = pd.read_csv('shopify/apps.csv', sep=',', header=0, lineterminator='\n')
apps_rows = df.values.tolist()

# embed key_benefits into apps
df = pd.read_csv('shopify/key_benefits.csv', sep=',', header=0, lineterminator='\n')
key_ben_rows = df.values.tolist()

i = 0
for app_row in apps_rows:
    
    record = {}
    record['id'] = app_row[0]
    record['url'] = app_row[1]
    record['title'] = app_row[2]
    record['developer'] = app_row[3]
    record['developer_link'] = app_row[4]
    record['icon'] = app_row[5]
    record['rating'] = app_row[6]
    record['reviews_count'] = app_row[7]
    
    apps_benefits = []
    
    for key_ben_row in key_ben_rows:
        if app_row[0] == key_ben_row[0]:

            app_benefit = {}
            app_benefit['title'] = key_ben_row[1]
            app_benefit['description'] = key_ben_row[2]
            
            apps_benefits.append(app_benefit)
    
    if len(apps_benefits) > 0:
        record['key_benefits'] = apps_benefits
    
    apps_ref = db.collection('apps').document(app_row[0])
    
    # populate the documents
    batch.set(apps_ref, record)
    if i == 100:
        batch.commit()
        i = 0
    i += 1
    
batch.commit()
```

```python
import time
batch = db.batch()

df = pd.read_csv('shopify/reviews.csv', sep=',', header=0, lineterminator='\n')
rows = df.values.tolist()

i = 0
for row in rows:
    record = {}
    record['app_id'] = row[0]
    record['author'] = row[1]
    record['rating'] = row[2]
    record['posted_at'] = row[3]
    
    unique_id = (row[0] + '_' + row[1]).replace(" ", "")
    
    reviews_ref = db.collection('/reviews').document(unique_id)
    
    batch.set(reviews_ref, record)
    
    if i >= 100:
        batch.commit()
        i = 0
        time.sleep()
    i += 1
    
batch.commit()
```

```python
# list 10 apps with highest number of reviews
```

```python
# need to get this to return the rows
apps_reviews = db.collection('apps')
query = apps_reviews.order_by('reviews_count', direction = firestore.Query.DESCENDING).limit(10)
results = query.stream()

for result in results:
    print(f'{result.id} => {result.to_dict()}')
```

```python
# apps categorized as trust and security
```

```python
category_ref = db.collection('categories')
query = category_ref.where('title', '==', 'Trust and security').limit(5)
results = query.stream()

title = None

for result in results:
    cat_record = result.to_dict()
    title = cat_record['title']

results = db.collection('apps').stream()

for result in results:
    app = result.to_dict()
    if 'one_app' in app.keys():
        one_app = app['one_app']
        
        for an_app in one_app:
            if title == an_app['title']:
                print(f'{result.id} => {result.to_dict()}')
                break
```
